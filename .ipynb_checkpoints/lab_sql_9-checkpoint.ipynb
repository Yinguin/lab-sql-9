```python
import pandas as pd
import numpy as np
import sqlalchemy as alch
from getpass import getpass
```

```python
dbName = "sakila"
connectionData=f"mysql+pymysql://root:{password}@localhost/{dbName}"
engine = alch.create_engine(connectionData)
```

```python
rentals_may = pd.read_sql_query('SELECT * FROM rentals_may', engine)
rentals_may.head()
```

```python
rentals_june = pd.read_sql_query('SELECT * FROM rentals_june', engine)
rentals_june.head()
```

```python
rentals_may_count = rentals_may.groupby('customer_id')['rental_id'].count().reset_index()
rentals_may_count = rentals_may_count.rename(columns={'rental_id': 'rentals_may'})
rentals_may_count
```

```python
rentals_june_count = rentals_june.groupby('customer_id')['rental_id'].count().reset_index()
rentals_june_count = rentals_june_count.rename(columns={'rental_id': 'rentals_june'})
rentals_june_count
```

```python
merge_data = pd.merge(rentals_may_count, rentals_june_count, on='customer_id', how='outer').sort_values('customer_id').reset_index(drop=True)
merge_data.fillna(0, inplace=True)
merge_data
```

```python
merge_data['compare'] = np.where(merge_data['rentals_june'] > merge_data['rentals_may'], 'more', 
                                 np.where(merge_data['rentals_june'] < merge_data['rentals_may'], 'less', 'same'))
merge_data
```

```python
def compare_rentals(rentals_may, rentals_june):
    # Group by customer ID and count number of rentals in May
    rentals_may_count = rentals_may.groupby('customer_id')['rental_id'].count().reset_index()
    rentals_may_count = rentals_may_count.rename(columns={'rental_id': 'rentals_may'})
    
    # Group by customer ID and count number of rentals in June
    rentals_june_count = rentals_june.groupby('customer_id')['rental_id'].count().reset_index()
    rentals_june_count = rentals_june_count.rename(columns={'rental_id': 'rentals_june'})
    
    # Merge the two dataframes on customer ID
    merge_data = pd.merge(rentals_may_count, rentals_june_count, on='customer_id', how='outer')
    
    # Fill any missing values with 0
    merge_data = merge_data.fillna(0)
    
    # Add a column to compare the number of rentals in May vs June
    merge_data['compare'] = np.where(merge_data['rentals_june'] > merge_data['rentals_may'], 'more', np.where(merge_data['rentals_june'] < merge_data['rentals_may'], 'less', 'same'))
    
    # Return the merged dataframe
    return merge_data
```

```python
compare_rentals(rentals_may, rentals_june)
```
